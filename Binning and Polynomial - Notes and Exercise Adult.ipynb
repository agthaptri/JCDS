## Binning with Sklearn

```python
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
disk = pd.read_csv('disk.csv')
disk.head()
```

```python
sns.kdeplot(data=disk['x'])
# dist tidak normal
```

```python
sns.scatterplot(x='x', y='class', data=disk)
```

```python
sns.barplot(x='class',y='x',data=disk)
```

```python
disk['class'].value_counts()/disk.shape[0]*100
# sekilas tidak ada pattern apa2
```

Mari kita coba lakukan binning variabel 'x' dan hitung proporsi dari tiap kelas

```python
disk['x eqint'] = pd.cut(disk['x'],7) #karena dist tidak normal
disk['x eqint encode'] = pd.cut(disk['x'],7, labels=range(1,8))
```

```python
disk
```

```python
pd.crosstab(disk['x eqint'], disk['class'], normalize=0)
```

```python
pd.crosstab(disk['x eqint'], disk['class'], normalize=0).plot(kind='bar', stacked=True)
```

Di ilustrasi di atas, kita bisa melihat proporsi class dari 7 kategori binning data x. Proporsi class 1 menurun sampai bin ke 4, lalu meningkat setelahnya. Kita melihat ada hubungan non linear (kuadratik) antara x dan proporsi class. Penambahan fitur polinomial bisa meningkatkan performa prediksi, tapi di contoh ini, kita akan coba dengan binning.

### Model without binning

```python
x = disk[['x']] #harus pakai 2 siku
y = disk['class']

x_train, x_test, y_train, y_test = train_test_split(x,y, stratify=y, random_state=2020)
```

```python
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
y_pred = logreg.predict(x_test)
print('akuraso:', accuracy_score(y_test,y_pred)) #akurasi awal tanpa binning
```

### Model with binning

```python
discretize = KBinsDiscretizer(n_bins=7, encode='onehot', strategy='uniform') #uniform = eq int, quantile = eq freq
x_train_disc = discretize.fit_transform(x_train)
x_test_disc = discretize.transform(x_test)
```

```python
logreg = LogisticRegression()
logreg.fit(x_train_disc,y_train)
```

```python
y_pred = logreg.predict(x_test_disc)
print('akurasi:', accuracy_score(y_test, y_pred))
```

Setelah memakai binning, akurasi meningkat dari 58.22% menjadi 69.97%

# Polynomial

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
wine = pd.read_csv('white_wine.csv')
```

```python
wine.isna().sum()
```

```python
wine.head()
```

x = alcohol, density  
y = quality >6 = 1, else 0

```python
wine['alcohol'] = wine['alcohol'].fillna(wine['alcohol'].mean())
wine['label'] = np.where(wine['quality']>6, 1, 0)
```

```python
wine['label'].value_counts() # terlihat data timpang antara 1 dan 0
```

```python
x = wine[['alcohol','density']]
y = wine['label']
```

```python
# Splitting data
x_train, x_test, y_train, y_test = train_test_split(x,y, stratify=y, random_state=2020)
```

## Without polynomial

```python
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
y_pred =  logreg.predict(x_test)
print('akurasi:', accuracy_score(y_test, y_pred))
```

## With Polynomial

```python
poly = PolynomialFeatures(degree=3, interaction_only=False, include_bias=False)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.fit_transform(x_test)
```

```python
poly.get_feature_names()
# semua yang bisa fit transform ..... bisa digunakan untuk semua operasi encoding dll?
```

```python
pd.DataFrame(x_train_poly, columns=poly.get_feature_names())
```

```python
logreg = LogisticRegression(max_iter=1000)
logreg.fit(x_train_poly, y_train)
```

```python
y_pred = logreg.predict(x_test_poly)
print('akurasi:', accuracy_score(y_test, y_pred))
```

Akurasi model meningkat dari 85.38% menjadi 96.92% setelah menggunakan polynomial dengan degree 3

```python
df = pd.read_csv('adult.csv')
df.head()
```

```python
df['income'].value_counts()
```

### Missing Value
Missing value in this dataset marked with "?"

```python
df = df.replace('?', np.nan)
```

```python
df.isna().sum()
```

### Preprocessing

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

import category_encoders as ce
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
pipe_impute_binary = Pipeline([
    ('imputer',SimpleImputer(strategy = 'constant',fill_value = 'NA')),
    ('binary encoder',ce.BinaryEncoder())
])

transformer = ColumnTransformer([
    ('one hot encoder',OneHotEncoder(drop = 'first'),['relationship','race','sex']),
    ('binary encoder',pipe_impute_binary,['workclass','marital.status','occupation','native.country'])
], remainder = 'passthrough')
```

### Splitting Data

```python
x = df.drop(columns=['fnlwgt', 'income', 'education']) #education di drop karena sudah terwakili education.num
y = np.where(df['income']=='>50K', 1, 0)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, stratify=y, random_state=10)
```

### Data Tramsform

```python
x_train.head()
```

```python
x_train_preprocessed = transformer.fit_transform(x_train)
x_test_preprocessed = transformer.transform(x_test)
```

```python
x_train_preprocessed = pd.DataFrame(x_train_preprocessed)
x_test_preprocessed = pd.DataFrame(x_test_preprocessed)
```

```python
x_train_preprocessed
```

```python
transformer.transformers_
```

```python
transformer.transformers_[0][1].get_feature_names()
```

```python
transformer.transformers_[1][1]['binary encoder'].get_feature_names()
```

```python
features = list(transformer.transformers_[0][1].get_feature_names()) + transformer.transformers_[1][1]['binary encoder'].get_feature_names() + ['age','education.num','capital.gain','capital.loss','hours.per.week']
len(features)
```

```python
x_train_preprocessed.columns = features
x_test_preprocessed.columns = features
```

```python
x_train_preprocessed
```

### Model Fitting and Evaluation

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
```

```python
tree = DecisionTreeClassifier(max_depth = 5, criterion='entropy')
tree.fit(x_train_preprocessed, y_train)
```

```python
y_predict = tree.predict(x_test_preprocessed)
print('mse:', mean_squared_error(y_test,y_predict))
print('akurasi:',accuracy_score(y_test,y_predict))
```

### With Polynomial

```python
poly = PolynomialFeatures(degree=3, interaction_only=False, include_bias=False)
x_train_poly = poly.fit_transform(x_train_preprocessed)
x_test_poly = poly.fit_transform(x_test_preprocessed)
```

```python
len(poly.get_feature_names())
```

```python
pd.DataFrame(x_train_poly, columns=poly.get_feature_names())
```

```python
logreg = LogisticRegression(max_iter=100000)
logreg.fit(x_train_poly, y_train)
```

```python
y_pred = logreg.predict(x_test_poly)
print('akurasi:', accuracy_score(y_test, y_pred))
```
